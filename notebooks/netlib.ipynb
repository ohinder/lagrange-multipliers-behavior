```julia
include("../src/shared.jl")
max_it = 100;
max_problem_size = 10000
```

```julia
# calculate which Netlib LP problems have no interior 
problem_list = load_netlib(max_problem_size,"../netlib/")
tol_set = [1e-4, 1e-6, 1e-8]
probs_no_int_tol_set = problems_with_no_interior(problem_list, tol_set)
```

```julia
compare_across_tols(probs_no_int_tol_set, tol_set); # check tolerance doesn't really matter
problems_with_int = problem_list[probs_no_int_tol_set[tol_set[1]]]; # problems with interior
problems_with_no_int = problem_list[.!probs_no_int_tol_set[tol_set[1]]]; # problems with no interior
```

```julia
# compute number of failures for all three methods
restricted_problem_list = intersect(problem_list,problems_with_no_int)

solver_dic = Dict(
"One Phase" => OnePhase.OnePhaseSolver(term!max_it= max_it, output_level=0),
"Ipopt" => IpoptSolver(print_level=2, max_iter = max_it, bound_relax_factor=0.0, acceptable_iter=99999, acceptable_tol=Inf, acceptable_compl_inf_tol=Inf, acceptable_constr_viol_tol=Inf, nlp_scaling_method="none")
)

solver_status_dic = compute_solver_status(solver_dic, restricted_problem_list);
```

```julia
print_solvers_failures(solver_status_dic)
problems_where_all_solvers_succeed = all_solvers_succeed(solver_status_dic, restricted_problem_list);
```

```julia
# generate distribution on the **maximum** dual multiplier values over **all iterates**
solver_duals = get_maximum_duals_all_iterates(solver_dic, problems_where_all_solvers_succeed, frac=1.0)
solver_duals_10_percent = get_maximum_duals_all_iterates(solver_dic, problems_where_all_solvers_succeed,frac=0.1)
```

```julia
plot_duals(solver_duals)
```
