```julia
# calculate which Netlib LP problems have no interior 

using Gurobi

problem_list = load_netlib(99999999,"../netlib/")

tol_set = [1e-4, 1e-6, 1e-8]
status_list = Dict{Float64,Array{String,1}}()

for tol in tol_set
    status_list[tol] = Array{Float64,1}()
    for problem_name in problem_list
      LP = read_lp(problem_name,"../netlib/")
      string_status = lp_feasible(LP, tol)
      push!(status_list[tol], string_status)
    end
end

@show length(problem_list)
```

```julia
# compute number of failures for all three methods

solver_status = Dict{String,Dict{String,Symbol}}()
problem_list = load_netlib(10000,"../netlib/")

solver_dic = Dict(
"OnePhase" => OnePhase.OnePhaseSolver(term!max_it=100),
"Ipopt" => IpoptSolver(print_level=2, max_iter = 100, bound_relax_factor=0.0, acceptable_iter=99999, acceptable_tol=Inf, acceptable_compl_inf_tol=Inf, acceptable_constr_viol_tol=Inf, nlp_scaling_method="none")
)

for (solver_name,solver) in solver_dic
    solver_status[solver_name] = Dict{String,Symbol}()
    
    for problem_name in problem_list
      println("Solving $problem_name")
      LP = read_lp(problem_name,"../netlib/")
      m = build_LP_model_as_NLP(LP)
      setsolver(m, solver)
      status = solve(m)
      solver_status[solver_name][problem_name] = status
    end
end
```

```julia
println("===== Which problems do the solvers fail on? ======")

for (solver_name,solver) in solver_dic
    num_failures = 0
    println("==== Solver: ", solver_name, " ======")
    for (prob_name,status) in solver_status[solver_name]
        if status != :Optimal
            println(prob_name, " ", status)
            num_failures += 1;
        end
    end
    println("total failures = ",num_failures)
end
```

```julia
# construct a list of problems for which both solvers succeed
both_succeed = Array{String,1}();
for prob_name in problem_list
    succeed = true
    for (solver_name,solver) in solver_dic
        if solver_status[solver_name][prob_name] != :Optimal
            succeed = false
        end
    end
    if succeed == true
        push!(both_succeed,prob_name)
    end
end
```

```julia
solver_dic = Dict(
"OnePhase" => OnePhase.OnePhaseSolver(term!max_it=100),
"Ipopt" => IpoptSolver(print_level=2, max_iter = 100, bound_relax_factor=0.0, acceptable_iter=99999, acceptable_tol=Inf, acceptable_compl_inf_tol=Inf, acceptable_constr_viol_tol=Inf, nlp_scaling_method="none")
)

function get_mult_inf_norm(inn::OnePhase.OnePhaseProblem)
    return norm(inn.lambda,Inf)
end

function get_mult_inf_norm(inn::IpoptProblem)
    return max(norm(inn.mult_g,Inf),norm(inn.mult_x_L,Inf),norm(inn.mult_x_U,Inf))
end

# generate distribution on the **maximum** dual multiplier values at the **last iterate**
solver_duals = Dict{String,Dict{String,Float64}}()
for (solver_name,solver) in solver_dic
    solver_duals[solver_name] = Dict{String,Float64}()
    
    for problem_name in both_succeed
      println("Solving $problem_name")
      LP = read_lp(problem_name,"../netlib/")
      m = build_LP_model_as_NLP(LP)
      setsolver(m, solver)
      status = solve(m)
      solver_duals[solver_name][problem_name] = get_mult_inf_norm(m.internalModel.inner)
    end
end
```

```julia
duals_IP = collect(values(solver_duals["Ipopt"]));
duals_OP = collect(values(solver_duals["OnePhase"]));
bins = 1:12
histogram([log.(duals_IP)/log(10), log.(duals_OP)/log(10)],bins=bins,fillalpha=0.8)
```
